**#Question 1: What is Simple Linear Regression?**

Simple linear regression is a statistical method used to find and describe the relationship between **two variables** by fitting a straight line to the data.

Think of it as a way to predict one thing (the **dependent variable**) based on another thing (the **independent variable**).

### Core Components

* **Independent Variable (X):** The factor you think causes a change (e.g., hours studied).
* **Dependent Variable (Y):** The factor you want to predict (e.g., test score).
* **The Best-Fit Line:** A straight line drawn through data points that minimizes the distance between the line and the actual points.

### How It Works

The relationship is expressed as a simple mathematical formula:

* \beta_0 (**Intercept**): Where the line hits the vertical axis (the value of Y when X is zero).
* \beta_1 (**Slope**): How much Y changes for every 1-unit increase in X.
* \epsilon (**Error**): The small difference between the actual data points and the line.

### A Real-World Example

Imagine you want to know how **Experience** (X) affects **Salary** (Y).

1. You collect data from several employees.
2. You plot these points on a graph (a scatter plot).
3. Simple linear regression finds the "best" straight line through those points.
4. **The Result:** You can now predict a person's salary just by knowing their years of experience!

**#Question 3: What is heteroscedasticity, and why is it important to address in regression models?**

**Heteroscedasticity** is a statistical term used when the "spread" (variance) of the errors in a regression model is **not constant**.

Instead of the data points being evenly distributed around the regression line, they might start close together and fan out as the values increase, often forming a **cone or funnel shape**.

### Why It’s Important to Address

While heteroscedasticity doesn't usually make your predictions "wrong" (the coefficients remain unbiased), it causes three major problems:

1. **Unreliable P-values:** It biases the standard errors. This means your p-values might be much smaller than they should be, leading you to think a variable is "significant" when it actually isn't.
2. **Inefficiency:** The model is no longer the "Best Linear Unbiased Estimator" (BLUE). It essentially "ignores" that some data points are more reliable than others, resulting in less precise estimates.
3. **Invalid Confidence Intervals:** Since the standard errors are wrong, the margins of error (confidence intervals) for your predictions will be misleading.

### How to Fix It

* **Transformations:** Take the log or square root of the dependent variable (Y) to stabilize the variance.
* **Weighted Least Squares (WLS):** Assign more weight to data points with smaller variance and less weight to those with high variance.
* **Robust Standard Errors:** Use "sandwich" estimators (like White’s standard errors) to fix the p-values without changing the model itself.

**#Question 4: What is Multiple Linear Regression?**
**Multiple Linear Regression (MLR)** is an extension of simple linear regression used to predict the outcome of a **single dependent variable** based on **two or more independent variables**.

While simple regression looks at how *one* thing (like study hours) affects an outcome, multiple regression looks at how a *combination* of things (like study hours, sleep, and prior grades) work together to influence that same outcome.

### The Mathematical Formula

Instead of a single line on a 2D graph, multiple regression creates a "plane" or "hyperplane" in a multi-dimensional space:

* **Y**: The value you are trying to predict.
* **\beta_0**: The intercept (the value of Y when all X variables are zero).
* **X_1, X_2, X_n**: The different independent variables (predictors).
* **\beta_1, \beta_2, \beta_n**: The coefficients for each variable. These represent how much Y changes for every one-unit increase in that specific X, **assuming all other variables stay the same.**
* **\epsilon**: The model's error term.

### Why Use It?

1. **Better Accuracy:** Real-world outcomes are rarely caused by just one factor. Including more variables usually leads to more precise predictions.
2. **Control for Other Factors:** It allows you to see the effect of one variable while "holding constant" the others. For example, you can see the impact of exercise on weight loss while controlling for calorie intake.

### Key Example

If you are predicting **House Price** (Y):

* **X_1**: Square footage
* **X_2**: Number of bedrooms
* **X_3**: Distance from the city center

The model will tell you exactly how much each additional bedroom adds to the price, regardless of the square footage.

**#Question 5: What is polynomial regression, and how does it differ from linear regression?**
**Polynomial Regression** is a form of regression analysis where the relationship between the independent variable X and the dependent variable Y is modeled as an **n^{th} degree polynomial**.

While it fits a curved line to the data, it is technically still considered "linear" regression because the coefficients (\beta) are linear, even though the input features (like X^2 or X^3) are not.

### Key Differences

| Feature | Linear Regression | Polynomial Regression |
| --- | --- | --- |
| **Shape** | Fits a **straight line**. | Fits a **curved line**. |
| **Complexity** | Simple; low risk of overfitting. | Complex; high risk of **overfitting**. |
| **Equation** | Y = \beta_0 + \beta_1 X | Y = \beta_0 + \beta_1 X + \beta_2 X^2 + ... |
| **Best Use Case** | When the trend is consistent and direct. | When data has peaks, valleys, or curves. |

### Why use it instead of Linear Regression?

Linear regression will fail (underfit) if your data follows a curve—for example, if you are measuring the trajectory of a ball or the growth rate of bacteria. Polynomial regression adds "flexibility" by allowing the model to bend and follow those non-linear patterns.

```python
''' Question 6: Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
● X = [1, 2, 3, 4, 5]
● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points. '''
import numpy as np

X = np.array([1, 2, 3, 4, 5])
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Means of X and Y
x_mean = np.mean(X)
y_mean = np.mean(Y)

# Calculate Slope (beta_1)
numerator = np.sum((X - x_mean) * (Y - y_mean))
denominator = np.sum((X - x_mean)**2)
beta_1 = numerator / denominator

# Calculate Intercept (beta_0)
beta_0 = y_mean - (beta_1 * x_mean)

print(f"Manual Slope: {beta_1:.2f}")
print(f"Manual Intercept: {beta_0:.2f}")

```

```python
''' Question 7: Fit a Multiple Linear Regression model on this sample data:
● Area = [1200, 1500, 1800, 2000]
● Rooms = [2, 3, 3, 4]
● Price = [250000, 300000, 320000, 370000]
Check for multicollinearity using VIF and report the results. '''
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 1. Prepare the Data
data = pd.DataFrame({
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
})

X = data[['Area', 'Rooms']]
y = data['Price']

# 2. Fit Multiple Linear Regression
model = LinearRegression()
model.fit(X, y)

# 3. Calculate VIF for Multicollinearity
# We add a constant (intercept) for VIF calculation as required by statsmodels
X_vif = data[['Area', 'Rooms']]
vif_data = pd.DataFrame()
vif_data["Feature"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(len(X_vif.columns))]

# 4. Results
print("--- Model Coefficients ---")
print(f"Intercept: {model.intercept_:.2f}")
print(f"Coefficients: {dict(zip(X.columns, model.coef_.round(2)))}")
print("\n--- Multicollinearity Check (VIF) ---")
print(vif_data)
```

```python
''' Question 8: Implement polynomial regression on the following data:
● X = [1, 2, 3, 4, 5]
● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
Fit a 2nd-degree polynomial and plot the resulting curve. '''
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# 1. Prepare the data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# 2. Transform the data for a 2nd-degree polynomial (X^2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# 3. Fit the model
model = LinearRegression()
model.fit(X_poly, Y)

# 4. Create a smooth curve for plotting
X_range = np.linspace(1, 5, 100).reshape(-1, 1)
Y_curve = model.predict(poly.transform(X_range))

# 5. Plotting
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X_range, Y_curve, color='green', label='2nd Degree Polynomial')
plt.title('Polynomial Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Output the equation
intercept = model.intercept_
coef = model.coef_
print(f"Equation: Y = {intercept:.2f} + {coef[1]:.2f}X + {coef[2]:.2f}X²")

```

```python
''' Question 9: Create a residuals plot for a regression model trained on this data:
● X = [10, 20, 30, 40, 50]
● Y = [15, 35, 40, 50, 65]
Assess heteroscedasticity by examining the spread of residuals. '''
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Prepare data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# 2. Fit the model and get predictions
model = LinearRegression()
model.fit(X, Y)
Y_pred = model.predict(X)

# 3. Calculate residuals
residuals = Y - Y_pred

# 4. Plotting residuals
plt.scatter(Y_pred, residuals, color='blue', edgecolors='black')
plt.axhline(y=0, color='red', linestyle='--') # Zero-error line
plt.title('Residuals Plot')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()

```

**#Question 10: Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location.However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model.**

 In a real-world scenario like real estate pricing, variables like **Area** and **Rooms** almost always move together (multicollinearity), and expensive houses often have much higher price fluctuations than cheaper ones (heteroscedasticity).

---

### Step 1: Address Multicollinearity (Feature Redundancy)

Multicollinearity makes it hard to determine which feature is actually driving the price.

* **Calculate VIF:** I would calculate the **Variance Inflation Factor** for all features. Any feature with a VIF > 5 or 10 is a candidate for removal.
* **Feature Engineering:** Instead of using both "Number of Rooms" and "Total Area," I might combine them into a single feature like **"Average Room Size"** or simply drop the one that is less correlated with price.
* **Principal Component Analysis (PCA):** If I need to keep the information from all variables, I could use PCA to transform the correlated features into a set of uncorrelated "components."

### Step 2: Address Heteroscedasticity (Uneven Variance)

Real estate data often shows a "fan shape" where error increases as house price increases.

* **Log Transformation:** This is the "gold standard" for house prices. By taking the **Natural Log of Price**, I can compress the scale, making the variance more constant across the board.
* **Weighted Least Squares (WLS):** I would give more "weight" to data points where we have higher confidence (usually smaller, more common houses) and less weight to outliers (luxury mansions) to prevent them from skewing the model.
* **Use Robust Standard Errors:** I would apply **White’s standard errors** to ensure that my p-values and confidence intervals remain valid even if some heteroscedasticity remains.

### Step 3: Model Validation & Robustness

Once the data is cleaned and transformed:

* **Re-check Residuals:** I would recreate the residual plot to ensure the "fan shape" has disappeared.
* **Regularization (Ridge/Lasso):** I would consider using **Ridge Regression**. It is specifically designed to handle multicollinearity by adding a penalty to the coefficients, preventing them from becoming too large and unstable.
* **Cross-Validation:** I would use k-fold cross-validation to ensure the model performs well on new, unseen houses, not just the ones in our current database.
